### Find out the no of male & female employees in the company

*Hint: Use can use a condition to find out the no of 1s and 0s in gender and sum it*

```python
#considering Male as 1 and Female as 2 from the data set
Male=len(data[data.Gender==1])
Female=len(data[data.Gender==2])
print("number of male in the data set are {}\nnumber of females in the dataset are {}".format(Male,Female))
```

### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
#importing matplotlib for plotting purpose for the below questions
import matplotlib.pyplot as mat
IQR_data=data['Age'].quantile(.75)-data['Age'].quantile(.25)
print("the interquartile rage is "+str(IQR_data))
```
